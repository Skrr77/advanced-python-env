```python
import matplotlib.pyplot as plt
import pandas as pd

# [cite_start]Сначала нужно создать данные (df), иначе графику нечего рисовать [cite: 55, 56]
data = {
    'Name': ['Alice', 'Bob', 'ANNA', 'David', 'Eva'],
    'Age': [19, 22, 21, 18, 23],
    'Score': [85, 78, 92, 65, 88]
}
df = pd.DataFrame(data)

# [cite_start]Теперь рисуем график [cite: 66, 67, 68]
plt.plot(df['Name'], df['Score'])
plt.xlabel('Student Name')
plt.ylabel('Score')
plt.title('Student Scores Line Plot')
plt.show()
```

```python
plt.bar(df['Name'], df['Score'])
plt.xlabel('Name')
plt.ylabel('Score')
plt.title('Student Names vs Scores')
plt.show()
```
